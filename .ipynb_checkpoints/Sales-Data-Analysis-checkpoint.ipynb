## Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Read the CSV File

```python
sales = pd.read_csv("data/sales-data.csv")
```

Let's see our DataFrame

```python
sales.head(10)
```

We are going to suppose that the Cur Total column is USD. Now we must format this column to get more easy to visualize

```python
sales["Cur Total"] = sales["Cur Total"].astype(float).round(2)
```

Let's see if what i did worked out

```python
sales.head()
```

Now i can view better our Data Frame

```python
def verify_null(dataframe):
    """Verify if a DataFrame is null"""
    is_null = dataframe.isnull().any().any()

    if is_null:
        return True
    return False
```

```python
verify_null(sales)
```

Using the function i created above we can verify if any DataFrame has Null values or not within it

After doing it we can create a column that represents the total sales

```python
sales_sorted = sales.sort_values(by="Transaction Date")
sales_sorted["Total Sales"] = sales_sorted["Cur Total"].cumsum()
sales_sorted.head()
```

We got our sales Data Frame sorted and called it as sales_sorted so we preserv out original Data Frame from
being incorrectly edited. Then we made a new column called Total Sales that sum across each row from Cur Total column.

## Visualizing Data

See the currency we receive the most from our customers

```python
fig, ax = plt.subplots()
grouped_sales = sales_sorted.groupby("Currency")["Cur Total"].sum()
bar = ax.bar(x=grouped_sales.index, height=grouped_sales.values)
ax.ticklabel_format(useOffset=False, style="plain", axis="y")
ax.set(title="Total Sales for each Currency",
       xlabel="Currency",
       ylabel="Total ($)");
```

Looking for the chart above we can certainly say that our business should invest in USA, Nigeria and United Kingdom.

```python
total_sales_for_date = sales_sorted.groupby(["Transaction Date", "Products"])["Cur Total"].sum()
total_sales_df = total_sales_for_date.unstack()
total_sales_df.plot(kind="bar", figsize=(30,20))
plt.title("Total Sell for Date & Products")
plt.xlabel("Date")
plt.ylabel("Total Sell ($)")
plt.savefig(");
```

Looking for the chart above we can assume that our only product that is consistent is Data Analysis Made Easy.

The 12 Strategies and Techies Guide are inconsistent, and we should improve the product quality or the marketing or cancel
these products

Now we are going to see if our sells are increasing or not across time.

```python
fig, ax = plt.subplots(figsize=(30, 20))
grouped_sales_date = sales_sorted.groupby("Transaction Date")["Cur Total"].sum()
bar = ax.bar(x=grouped_sales_date.index, height=grouped_sales_date.values)
mean_sells = sales_sorted["Cur Total"].mean()
plt.axhline(mean_sells,
            linestyle=":",
            color="red")
plt.xticks(rotation=45);
plt.savefig("img/date-vs-total-sales.png")
```

The chart above tells us that we got a successful period between 04/29/2023 (a record) and 06/01/2023, then we got back to our average total sales but kept above the mean
